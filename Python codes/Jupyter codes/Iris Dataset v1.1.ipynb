```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
#print(iris_dataset)
```

```python
print(iris_dataset.keys())
print(iris_dataset["feature_names"])
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(iris_dataset["data"],iris_dataset["target"],random_state=0,test_size = 0.35)
print("X Training data set Type: {} and Shape: {}".format(type(xTrain),xTrain.shape))
print("Y Training data set Type: {} and Shape: {}".format(type(yTrain),yTrain.shape))
print("X Test data set Type: {} and Shape: {}".format(type(xTest),xTest.shape))
print("Y Test data set Type: {} and Shape: {}".format(type(yTest),yTest.shape))
```

```python
import pandas as pd
iris_dataframe = pd.DataFrame(data = xTrain, columns = iris_dataset["feature_names"])
#print(iris_dataframe)
grr = pd.plotting.scatter_matrix(iris_dataframe, c=yTrain, figsize=(10,10), marker="o",alpha=.8,s=40,hist_kwds={"bins":20})
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(xTrain,yTrain)
```

```python
import numpy as np
x_newdata = [5.9,1.2,1,.5]
print(x_newdata,"\n",type(x_newdata))
x_new_data = np.array([x_newdata])
print(x_new_data,"\n",type(x_new_data),x_new_data.shape)
```

```python
prediction = knn.predict(x_new_data)
print("Prediction: {}".format(prediction))
print("Predicted Target name: {}".format(iris_dataset["target_names"][prediction]))
```

```python
y_predict = knn.predict(xTest)
print("Test set prediction: {}".format(y_predict))
```

```python
print("Test set score: {:.2f}".format(np.mean(y_predict == yTest)))
print("KNN Test score: {:.2f}".format(knn.score(xTest,yTest)))
```
